```python
url = "https://api.simplyrets.com/properties"

for x in range (150):
    offset = 500*x 
    params = {'limit': '500', 
              'offset': offset}
    response = requests.get(url, params=params, auth=(api_key, api_secret)).json()
    response = json_normalize(response)
    df = pd.DataFrame(response)
    df_t = pd.concat([df_t, df], ignore_index=True, sort=False)
df_t
```

```python
df_t.to_csv("data_grabbed.csv", index=False, header=True)
```
