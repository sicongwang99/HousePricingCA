```python
import pandas as pd
```

```python
df = pd.read_csv('data_cleaned_final.csv')
df.head()
```

```python
# Only California State
df = df[['Remarks']]
df.head()
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df['rate'] = "string"
```

```python
for i, r in df.iterrows():
    target_string = df['Remarks'][i]
    results = analyzer.polarity_scores(target_string)
    compound = results["compound"]
    if compound >= 0.9:      
        df['rate'][i] = 'veryPositive'
    else:
        df['rate'][i] = 'positiveOrNegative'
```

```python
df.to_csv("compounded.csv", index=False, header=True)
```
