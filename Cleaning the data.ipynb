```python
import pandas as pd
```

```python
df = pd.read_csv('data_grabbed.csv')
df
```

```python
# Only California State
df = df.loc[:][df['address.state'] == "CA"].reset_index(drop=True)
df
```

```python
reduced_df = df[['address.city', 'address.full', 'address.postalCode', 'association.amenities', 'association.fee', 
                 'geo.lat', 'geo.lng', 'listDate', 'listPrice', 'mls.daysOnMarket', 'mls.status', 'photos','property.acres', 
                 'property.bathrooms', 'property.bedrooms', 'property.garageSpaces', 'property.lotSizeArea', 
                 'property.lotSizeAreaUnits', 'property.parking.spaces', 'property.subType', 'property.yearBuilt', 
                 'remarks', 
#                  'sales'
                ]]
reduced_df
```

```python
unique_reduced_df = reduced_df[['address.city', 'address.full', 'address.postalCode', 'association.amenities', 'association.fee', 
                 'geo.lat', 'geo.lng', 'listDate', 'listPrice', 'mls.daysOnMarket', 'mls.status', 'photos','property.acres', 
                 'property.bathrooms', 'property.bedrooms', 'property.garageSpaces', 'property.lotSizeArea', 
                 'property.lotSizeAreaUnits', 'property.parking.spaces', 'property.subType', 'property.yearBuilt', 
                 'remarks', 
#                  'sales'
                 ]]
unique_reduced_df = unique_reduced_df.drop_duplicates(keep='first').reset_index(drop=True)
unique_reduced_df
```

```python
reduced_na_df = unique_reduced_df[['address.city', 'address.full', 'address.postalCode', 'association.amenities', 'association.fee', 
                 'geo.lat', 'geo.lng', 'listDate', 'listPrice', 'mls.daysOnMarket', 'mls.status', 'photos','property.acres', 
                 'property.bathrooms', 'property.bedrooms', 'property.garageSpaces', 'property.lotSizeArea', 
                 'property.lotSizeAreaUnits', 'property.parking.spaces', 'property.subType', 'property.yearBuilt', 
                 'remarks',
#                  'sales'
                 ]].dropna(how='any')
reduced_na_df
```

```python
renamed_df = reduced_na_df.rename(columns={'address.city': 'City', 
                                        'address.full': 'Full_Address', 
                                        'address.postalCode': 'Postal_Code',
                                        'association.amenities': 'Amenities',
                                        'association.fee': 'Association_fee', 
                                        'geo.lat': 'latitude', 
                                        'geo.lng': 'longitude', 
                                        'listDate': 'List_Date', 
                                        'listPrice': 'List_Price', 
                                        'mls.daysOnMarket': 'MLS_Days_On_Market', 
                                        'mls.status': 'MLS_Status', 
                                        'photos': 'Photos', 
                                        'property.acres': 'Acres', 
                                        'property.bathrooms': 'Bathrooms', 
                                        'property.bedrooms': 'Bedrooms', 
                                        'property.garageSpaces': 'Garage_Spaces', 
                                        'property.lotSizeArea': 'Lot_Size_Area', 
                                        'property.lotSizeAreaUnits': 'Lot_Size_Area_Units', 
                                        'property.parking.spaces': 'Parking_Spaces', 
                                        'property.subType': 'Sub_Type', 
                                        'property.yearBuilt': 'Year_Built', 
                                        'remarks': 'Remarks', 
#                                         'sales':'Sales'
                                          }).reset_index(drop=True)

renamed_df
```

```python
cleaned_price = renamed_df.loc[:][renamed_df['List_Price'] > 100000].reset_index(drop=True)
cleaned_price = renamed_df.loc[:][renamed_df['List_Price'] < 3000000].reset_index(drop=True)
cleaned_price
```

```python
mask_remove_postalcode = (cleaned_price['Postal_Code'].astype(int) > 90000)
# mask_remove_postalcode.head()
# mask_remove_postalcode.sum()
postal_code = cleaned_price.loc[mask_remove_postalcode]
postal_code.shape
```

```python
cleaned_units = postal_code.loc[:][postal_code['Lot_Size_Area_Units'] == "Square Feet"]

cleaned_units = cleaned_units[['City', 'Full_Address', 'Postal_Code', 'Amenities', 'Association_fee', 'latitude', 
                               'longitude', 'List_Date', 'List_Price', 'MLS_Days_On_Market', 'MLS_Status', 'Photos', 
                               'Acres', 'Bathrooms', 'Bedrooms', 'Garage_Spaces', 'Lot_Size_Area', 'Parking_Spaces', 
                               'Sub_Type', 'Year_Built', 'Remarks'
#                                , 'Sales'
                              ]]
cleaned_units
```

```python
cleaned_photos = cleaned_units.loc[:][cleaned_units['Photos'] != "[]"].reset_index(drop=True)
cleaned_photos
```

```python
sort_by_year_built = cleaned_photos.sort_values("Year_Built", ascending=False).reset_index(drop=True)
sort_by_year_built
```

```python
cleaned_assiciated_fee = sort_by_year_built.loc[:][sort_by_year_built['Association_fee'] != 0.0].reset_index(drop=True)
cleaned_assiciated_fee
```

```python
# Keep the number equal to the above dataframe

for x in range(11417):
    cleaned_assiciated_fee['Photos'][x] = cleaned_assiciated_fee['Photos'][x].split(',')[0][2:-1]
cleaned_assiciated_fee
```

```python
cleaned_assiciated_fee.to_csv("data_cleaned_final.csv", index=False, header=True)
```
